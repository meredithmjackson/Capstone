```python
# load libraries
import json
from matplotlib import pyplot as plt
import numpy as np
```

```python
# load data
f = open(r'/Users/meredithjackson/Desktop/ga/capstone/data/shipsnet.json')
dataset = json.load(f)
f.close()

# show keys
dataset.keys()
```

```python
# transpose data, plot 0th image
data = np.array(dataset['data']).astype('uint8')
img_length = 80
data = data.reshape(-1,3,img_length,img_length).transpose([0,2,3,1])

plt.imshow(data[0])
```
